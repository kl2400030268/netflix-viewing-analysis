```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('netflix_viewing_data.xlsx')
df.head()
```

how many rows and columns

```python
df.shape
```

data types,missing values

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

which genre is watched the most?

```python
df['genre'].value_counts()
```

Drama,Action and Comedies are the most watched genre among all.

Average watch time by genre

```python
df.groupby('genre')['watch_time_hours'].mean()
```

Device usage distribution

```python
df['device'].value_counts()
```

TV is the most used.

Ratings by subscription type

```python
df.groupby('subscription_type')['rating'].mean()
```

Premium has the most rating out of all.

Bar chart

```python
df['genre'].value_counts().plot(kind='bar')
plt.title("Most Watched Genres")
plt.show()
```

pie chart

```python
df['device'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Device Usage")
plt.ylabel("")
plt.show()
```

line/bar chart

```python
df.groupby('age')['watch_time_hours'].mean().plot()
plt.title("Age vs Average Watch Time")
plt.show()
```

Conclusion:
Drama and Sci-Fi are the most watched genres

TV is the most commonly used device

Premium users spend more watch time on average

Younger users tend to watch more hours
